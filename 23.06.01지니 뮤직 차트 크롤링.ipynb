```python
import requests

# HTML, XML 에서 원하는데이터를 추출하게 해주는 라이브러리
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
def genie_chart():
    chart = []
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '}
    for i in range(1, 5) :
        url = f'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230601&hh=13&rtm=Y&pg={i}'
        req = requests.get(url, headers = header)
        soup = BeautifulSoup(req.text)
        info_td = soup.findAll('td', {'class':'info'})

        for j in range(len(info_td)):
            title = info_td[j].find('a').text.strip()
            artist1 = info_td[j].find('a', {'class','artist'})
            artist = artist1.text.strip()
            rank = (j+1)+((i-1)* 50)
            # chart.append('{0:3}위 {1} - {2}'.format((j+1)+((i-1)* 50),artist,title))
            dic = {'순위':rank, '가수':artist,'제목':title}
            chart.append(dic)
    return chart

df = pd.DataFrame(genie_chart())
```
